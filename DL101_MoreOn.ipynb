#Debugging

```
def factorial(x):
  if (x==0): # this creates a problem when the value is a negative number. It should be rectified . 
  #But for debugging practice, keeping as is
    return 1
  return x* factorial(x-1)
```

```
import random
def code_to_debug():
  import pdb;pdb.set_trace()
  for i in range(10) :
    x=random.random()
    factorial(x)
```

```
code_to_debug()
```

```
def factorial_debugged(x):
  if(not isinstance(x,int)):
    print('This method supports only integers')
    return -1
  if (x==0): 
    return 1
  return x* factorial_debugged(x-1)
```

```
def code_to_debug():
  import pdb;pdb.set_trace()
  for i in range(10) :
    x=random.random()
    factorial_debugged(x)
```
